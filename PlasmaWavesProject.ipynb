## Introduction: What is Plasma? 

In a general sense, a plasma is any matter that is at least partially ionized.$^1$

In other words, any matter that constitutes free charges - either positive or negative - is a plasma. A common and 'simple' description of plasma is a hydrodynamics one. In this model, a continuum description is used, such that instead of modeling each individual atom, a mean particles motion description is used. This involves parameters such as mean velocity and mass density. 

The continuum approximation is a very good one for plasma, since in general the mean free path of particles in plasma is very small (plasma is high density). This is the description that we will use to explain the propagations of waves in 2 specific types of plasmas: Electron and Ion. 

The goal of this work is to provide an interactive way for the user to study the formation of waves and oscillations in plasma once a perturbation, that results in the displacement of charged particle in the plasma, is applied.

## Electron Plasma Oscillations

Conceptual Idea: 

If an electron in a plasma (consisting of background of uniformly distributed ions) is perturbed, so as to move it away from a state of neutrality, the system will seek to restablish the neutrality. To do this, the electric fields will pull the electron back to its initial spot. However, because of the inertia of the electron, it will go past its initial spot, and begin oscillating around this equlibrium position with an angular _plasma frequency_ $\omega _p $

Finding $\omega _p $: 

Under the following simplifying conditions, 
   (1) No Magnetic Field
   (2) No Thermal Motion
   (3) Background Ions are Fixed in Place (and as Mentioned are Uniformly Distributed)
   (4) The Plasma System is Infinitly Large in all Directions
   (5) The Electrons Move in Only One Direction
 
,the equation of motion for an electron is: 

$\qquad  \qquad mn _e (\frac{\partial \boldsymbol{v _e}}{\partial t} + (\boldsymbol{v _e} \cdot \boldsymbol{\nabla})) = -en _e \textbf{E} \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad  \quad \quad(1)$

And the equation of continuity(satifies the conservation of matter for when there is a change of particles in some volume by taking into accout a flux of particles over the surface bounding the volume) is: 

$\qquad  \qquad \frac{\partial n _e}{\partial t} + \boldsymbol{\nabla} \cdot (n _e \boldsymbol{v _e}) = 0  \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad  \qquad \qquad \quad (2)$   

where $n _e$ is the electron plasma density and $\boldsymbol{v _e}$ is the electron velocity.

$ \omega _p $ can then be derived to be: $^2$ 

$ \qquad  \qquad \omega _p  = (\frac{n _e e^2 }{	\epsilon _0 m})^{1/2} \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad  \qquad \qquad \qquad \qquad (3)$

where e is the charge of an electron and m is the mass of the electron. 

Then we can approximate the to be: $^3$  
$ \qquad  \qquad \omega _p / 2\pi  = 9n^{1/2} \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad  \qquad \qquad \qquad \qquad (4)$

## Electron Plasma Waves

Conceptual Idea: 

If electrons in plasma move with thermal velocity into a certain region of the plasma that is experiencing an oscillation, they will carry with it information about the oscillation. Thus, in this situation, the plasma oscillation becomes a plasma _wave_.

To deal with this new component of motion, we need to add a term to represent the fluid nature of the motion to equation (1). This term is $\boldsymbol{\nabla} p _e = 3KT _e\boldsymbol{\nabla}n _e$ . Where $\boldsymbol{\nabla} p _e$ is the pressure gradient (for more information see Navier Stokes Equation), K is the Boltzmann constant, and $T _e$ is the temperature of the electrons.

From there we can derive the modified angular frequency as: $^4$

$ \qquad  \qquad \omega^2  = \omega _p ^2 + \frac{3}{2}k^2v _{th} ^2 \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad  \qquad \qquad \qquad \qquad (5)$

where $v _{th}$ is the thermal velocity of the electrons, and $v _{th}^2 \equiv 2KT _e/m $

In the code below, the user will be able to vary the plasma density and KT eV, and see the resulting plot of angular frequency of the plasma wave vs k. Note the velocity of the wave is given by $\omega / k$ (the slope of the curve).

$^{[4]: The-detailed-derivation-can-be-found-in:-Introduction-to-Plasma-Physics-and-Controlled-Fusion-by-F.F.-Chen-page-88}$



```python
# In this code, the user will be able to vary the plasma density and KT eV, and see the resulting plot of angular frequency of 
# the plasma wave vs k


import ipywidgets as widgets
from ipywidgets import HBox, VBox
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline

@widgets.interact_manual(
    color=['blue', 'red', 'green'], lw=(1., 10.))
def plot(KT=100.,n = 10**5., color='blue', lw=2., grid=True):
    k = np.linspace(0.0, 10., 1000) # all the k values
    #n = 10
    # the numbers from this calculation come from taking KT to be in eV. so 1.76*10^11 is equal
    # to 1 ev / m
    # The user can vary KT in eV
    #v_th_squared = 2*1.76*10^11 *KT 
    #omega_p = 9*np.sqrt(n) * 2* np.pi
    
    #omega_squared = omega_p^2 + k^2*v_th_squared * 3/2
    
    #((9*np.sqrt(n)*2*np.pi)^2 + k^2*(2*1.76*10^11 *KT)*3/2)
    
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    ax.plot(k,np.sqrt(((9*np.sqrt(n)*2*np.pi)**2)+ k*k*(2*KT)*3/2) ,lw=lw, color=color)
    ax.grid(grid)
    plt.xlabel('$k$');
    plt.ylabel('$\omega $')
```

The formula for the electron plasma density *wave* then is: 

$ n(x,t) = n _e cos(kx- \omega t) $

```python
#In the code below, the user is able to change the electron density, $n _e$, in the plasma, k, and KT_e eV, 
#to see how the wave of electron density changes in a static case (for time t = 1).

import ipywidgets as widgets
from ipywidgets import HBox, VBox
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline

@widgets.interact_manual(
    color=['blue', 'red', 'green'], lw=(1., 10.))
def plot(n_e = 10**5,k = 1., KT_e=100., color='blue', lw=2., grid=True):
    x = np.linspace(0.0, 30., 1000) # all the k values
    t = 1
    omega_p = 9*np.sqrt(n_e) * 2*np.pi
    vth = 2*KT_e*1.76*(10**11)
    omega = np.sqrt(omega_p**2 + 3/2*k*vth)
    # the numbers from this calculation come from taking KT to be in eV. so 1.76*10^11 is equal
    # to 1 ev / m
    # The user can vary KT in eV
    #v_th_squared = 2*1.76*10^11 *KT 
    #omega_p = 9*np.sqrt(n) * 2* np.pi
    
    #omega_squared = omega_p^2 + k^2*v_th_squared * 3/2
    
    #((9*np.sqrt(n)*2*np.pi)^2 + k^2*(2*1.76*10^11 *KT)*3/2)
    
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    ax.plot(x,n_e*np.cos(k* x + omega*t) ,lw=lw, color=color)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$n$ - Electron Density')
    ax.grid(grid)
```

## Ion Plasma Waves

For ion plasmas, we now enter a regime of a much larger scale than we were previous considering with electrons. As such, ion plasma waves can be modeled by sound waves (a common example of course is that of air molecules).  

For a general sound wave in which we are neglecting viscosity and in which there are no electromagentic forces, we can write the Navier-Stokes equation as: 

$ \qquad  \qquad \rho (\frac{\partial \boldsymbol{v}}{\partial t} + (\boldsymbol{v} \cdot \boldsymbol{\nabla})\boldsymbol{v}) = -\boldsymbol{\nabla}p = -\frac{\gamma p}{\rho}\boldsymbol{\nabla}\rho \qquad \qquad \qquad \qquad \qquad  \qquad \qquad \qquad \qquad (6)$

Notice that in air molecule sound waves, information is transmitted from region to region via collisions of air molecules(allowing an oscillation to propagate). In order for ion plasma waves to propagate, this "limitation" is no longer in place, as oscillations can be transmitted from region to region via the coloumb forces due to the charges of the ions. 

Without a magnetic field, the sound waves description of ion plasma waves, with the plasma approximation(i.e. that the density of ions equals the density of electrons $\equiv$ n), is: 

$ \qquad  \qquad Mn(\frac{\partial \boldsymbol{v _i}}{\partial t} + (\boldsymbol{v _i} \cdot \boldsymbol{\nabla})\boldsymbol{v _i}) = en\boldsymbol{E}-\boldsymbol{\nabla}p = -en\boldsymbol{\nabla}\phi- \gamma _i K T _i\boldsymbol{\nabla}n  \qquad \qquad \qquad \qquad \qquad (7)$

Where M is the mass of the ions, $v _i$ is the velocity of the ions, $T _i$ is the temperature of the ions, and $\gamma _i$ is the ratio of the heat capacity of isobaric and isochoric processes.

Then the angular frequency of the wave can be derivied to be: $^5$

$\qquad  \qquad \omega = k \left(\frac{KT _e + \gamma _i KT _i}{M}\right)^\frac{1}{2} \qquad \qquad \qquad \qquad \qquad  \qquad \qquad \qquad \qquad \qquad \qquad \quad (8)$



$^{[5]: The-detailed-derivation-can-be-found-in:-Introduction-to-Plasma-Physics-and-Controlled-Fusion-by-F.F.-Chen- pages-96}$


```python
# In this code, the user will be able to vary the ion density, ion mass, KT_e eV and KT_i eV, and see the resulting plot of angular frequency of 
# the ion plasma acoustic wave vs k


import ipywidgets as widgets
from ipywidgets import HBox, VBox
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline

@widgets.interact_manual(
    color=['blue', 'red', 'green'], lw=(1., 10.))
def plot(KT_e=100.,KT_i = 10., M_exponent_scale = -11, color='blue', lw=2., grid=True):
    k = np.linspace(0.0, 10., 1000) # all the k values
    #gamma can be set to 3 commonly:
    gamma = 3
    #KT_e and KT_i are in eV: 1 eV = 1.602 * 10^-19
    
    #the equation for omega is:
    # omega = k*((KT_e + gamma*KT_i)*(1.602*10**(-19))/10**M_exponent_scale)**(1/2)
    
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    ax.plot(k,k*((KT_e + gamma*KT_i)*(1.602*10**(-19))/10**M_exponent_scale)**(1/2),lw=lw, color=color)
    ax.grid(grid)
    plt.xlabel('$k$');
    plt.ylabel('$\omega $');
    
```

# Comparing Electron Plasma and Ion Plasma Waves

The essential difference between waves generated in electron plasmas and ion plasmas is that: 
       (1) Electron waves are basically constant frequency
       (2) Ion waves are basically constant velocity 
       
The fundamental reason for this has do with the motion of the background species in the plasma. In electron waves the $\textit{ions}$ remained fixed, while in ion waves the $\textit{electrons}$ are experiencing thermal motions.

Since the electrons move along with ions, they provided a shielding of the electric field the ions generate, with only fields proptional to $KT_e$ making it through due to the thermal motions of electrons. This is the only electric field that acts between the clusters of ions that are formed. 

```python
# In this code, the user will be able to vary the plasma density and KT eV, and see the resulting plot of angular frequency of 
# the plasma wave vs k

import ipywidgets as widgets
from ipywidgets import HBox, VBox
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline

@widgets.interact_manual(
    color=['blue', 'red', 'green'], lw=(1., 10.), k_value_beg =(0,10000),k_value_end =(0,10000))
def plot(k_value_beg = 0,k_value_end = 10, color='blue', lw=2., grid=True):
    
    k = np.linspace(k_value_beg,k_value_end, 1000) # all the k values
    #n = 10
    # the numbers from this calculation come from taking KT to be in eV. so 1.76*10^11 is equal
    # to 1 ev / m
    # The user can vary KT in eV
    #v_th_squared = 2*1.76*10^11 *KT 
    #omega_p = 9*np.sqrt(n) * 2* np.pi
    
    #omega_squared = omega_p^2 + k^2*v_th_squared * 3/2
    
    #((9*np.sqrt(n)*2*np.pi)^2 + k^2*(2*1.76*10^11 *KT)*3/2)
    KT=100.
    n = 10**5.
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    ax.plot(k,np.sqrt(((9*np.sqrt(n)*2*np.pi)**2)+ k*k*(2*KT)*3/2) ,lw=lw, color=color)
    ax.grid(grid)
    plt.xlabel('k')
    plt.ylabel('$\omega$')
```

When deriving the frequency of the ion plasma waves, we made the approixmating assumption 
that $n_e = n_i$. This approximation can cause an error that hides valuable information when comparing the 2 types of plasmas. Because of this we will look at the what happens without this approximation.

Without this approximation, the equation for frequency dervives to be:$^6$

$\qquad  \qquad \omega = k \left(\frac{KT _e}{M}\frac{1}{1+k^2\lambda^2_D} + \frac{\gamma _i KT _i}{M}\right)^\frac{1}{2} \qquad \qquad \qquad \qquad \qquad  \qquad \qquad \qquad \qquad \qquad (10)$

As we can see, this equation is of the same form as equation (8), except for a factor of
$\frac{1}{1+k^2\lambda^2_D}$ multiplying the fist term. Thus, the regime where the approximation fails to provide us useful information is for waves of short wavelengths. I.e. in the limit: $k^2\lambda^2_D >> 1$. This is because in this regime, we no longer have 
$\frac{1}{1+k^2\lambda^2_D} \approx 1$. 

In this limit then, equation (10) becomes: 
$\qquad  \qquad \omega = k \left(\frac{KT _e}{M}\frac{1}{k^2\lambda^2_D} + \frac{\gamma _i KT _i}{M}\right)^\frac{1}{2} \qquad \qquad \qquad \qquad \qquad  \qquad \qquad \qquad \qquad \qquad (11)$

$^{[6]: The-detailed-derivation-can-be-found-in:-Introduction-to-Plasma-Physics-and-Controlled-Fusion-by-F.F.-Chen- pages-98}$

By plotting $ \omega $ for electron plasma, we can see that initially the electron is constant frequency, but for larger k, it becomes constant velocity (the graph below goes from being quadractic to being linearly).

  In this work we looked at the wave description of oscillations in electron plasmas and waves in both electron plasmas and ion plasmas. These oscillations and waves form when some type of perturbation is applied to a plasma so that a charged particle is displaced. For all these oscillations and waves, from the interactive simulations of the motion of the  density in plasma, the user was able to see which quantities control the amplitude, phase shift, and period of the waves(charged particle density at equilibrium, k, kinetic energy, etc). In additon, we compared the motions of electron and ion plasma waves. The conlusion of this comparioson was the following. For electron plasma, initially the electron is constant frequency, but for larger k, it becomes constant velocity. The opposite is true for ion plasma. Ion waves are initially constant velocity  ut become constant frequency for large k.
